```python
from googlesearch import search
from tqdm import tqdm
```

```python
CA_charters = open("California Charter Cities.txt",'r')
CA_charters = (CA_charters
                          .read()
                          .split('\n')
              )
```

```python
charter_urls = {}

for city in tqdm(CA_charters):
    while True:
        try:
            urls = [url for url in search(city + " california charter", tld="co.in", num=3, stop=3, pause=12)]
            charter_urls[city] = urls
            break
        except:
            print(f'retrying {city}')
            continue
    
```

```python
import json  
    
with open("CA_charters.json", "w") as outfile:  
    json.dump(charter_urls, outfile) 
```
